```python
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup
import requests
from six.moves.urllib import parse
import time
from selenium import webdriver
from bs4 import BeautifulSoup as bs
```

```python
# I used Firefox; you can use whichever browser you like.
browser = webdriver.Firefox()

# Tell Selenium to get the URL you're interested in.
browser.get("https://www.rfsafe.com/sar-values-specific-absorption-rate-comparison-database/")
```

```python
# Selenium script to scroll to the bottom, wait 3 seconds for the next batch of data to load, then continue scrolling.  It will continue to do this until the page stops loading new data.
lenOfPage = browser.execute_script("window.scrollTo(0, document.body.scrollHeight);var lenOfPage=document.body.scrollHeight;return lenOfPage;")
match=False   
while(match==False):
        lastCount = lenOfPage
        time.sleep(3)
        lenOfPage = browser.execute_script("window.scrollTo(0, document.body.scrollHeight);var lenOfPage=document.body.scrollHeight;return lenOfPage;")
        if lastCount==lenOfPage:
            match=True
# Now that the page is fully scrolled, grab the source code.
source_data = browser.page_source

# Throw your source into BeautifulSoup and start parsing!
bs_data = bs(source_data)

soup= bs_data
y=soup.prettify()
soup.findAll('a')

```

Collect each section of specification and put it into CSV file

```python
c = 1
d=[]

for li in soup.find_all('li'):
    for link in li.find_all('div',{'class':'aps-product-box'}):
        for link in li.find_all('h2',{'class':'aps-product-title'}):
            for anc in link.find_all('a'):
                anc_src = anc.get('href')
                anc_name = anc.string
                print(anc_src )
                d.append(anc_src)
                c += 1

df =pd.DataFrame(d)
df.replace('', np.nan, inplace=True)
print(df)
# Drop the empty cell
df1= df.dropna()

# Data is stored in CSV file
df1.to_csv("C:\\Users\\gaura\\Desktop\\Project\\link.csv", sep='\t', encoding='utf-8', index=False)

```
